```python
import matplotlib.pyplot as plt
import pandas as pd

# Funding data
years = [2016, 2018, 2021, 2023]
amounts_raised = [8, 20, 75, 30]  # in million USD
total_raised = 130  # total amount raised in million USD

# Cumulative funding over the years
cumulative_amounts = [sum(amounts_raised[:i+1]) for i in range(len(amounts_raised))]

# Create a dataframe for ease of plotting
data = pd.DataFrame({
    'Year': years,
    'Amount Raised': amounts_raised,
    'Cumulative Amount': cumulative_amounts
})

# Plotting the funding overview
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Cumulative Amount'], marker='o', color='lightseagreen', linewidth=3)
plt.fill_between(data['Year'], data['Cumulative Amount'], color='hotpink', alpha=0.3)
for i in range(len(data)):
    plt.text(data['Year'][i], data['Cumulative Amount'][i], f"${data['Cumulative Amount'][i]} Mn", 
             horizontalalignment='right' if i == len(data)-1 else 'center', fontsize=10, verticalalignment='bottom')
plt.title('HealthifyMe Funding Overview', fontsize=16)
plt.xlabel('Time Period (In Years)', fontsize=14)
plt.ylabel('Amount Raised (in USD)', fontsize=14)
plt.xticks(years)
plt.yticks(range(0, 90, 10))
plt.grid(True)
plt.text(2023, cumulative_amounts[-1], f"${total_raised} Mn+\nTotal Amount Raised", 
         horizontalalignment='left', verticalalignment='bottom', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Operating Revenue', 'Total Expenses', 'Profit/Loss', 'Cash from Ops']
fy21_values = [86.05, 105.82, -19, -4.12]
fy22_values = [185.25, 343.44, -157, -84.10]
growth_percentages = ['+115.3%', '+224.6%', '-726.3%', '-1941.3%']
growth_colors = ['green', 'green', 'red', 'red']

# Creating the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Bar width
bar_width = 0.35

# Position of bars on x-axis
r1 = range(len(fy21_values))
r2 = [x + bar_width for x in r1]

# Making the bar plot
bars1 = plt.bar(r1, fy21_values, color='lightseagreen', width=bar_width, edgecolor='grey', label='FY21')
bars2 = plt.bar(r2, fy22_values, color='hotpink', width=bar_width, edgecolor='grey', label='FY22')

# Adding value labels inside the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval - (0.1 * yval), round(yval, 2), ha='center', va='top', fontsize=10, fontweight='bold', color='black')
    
for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval - (0.1 * yval), round(yval, 2), ha='center', va='top', fontsize=10, fontweight='bold', color='black')

# Adding growth percentages above the bars
for i, (percentage, color) in enumerate(zip(growth_percentages, growth_colors)):
    if '-' in percentage:
        plt.text(r2[i], fy22_values[i] - 30, percentage, ha='center', va='top', fontsize=12, fontweight='bold', color=color)
    else:
        plt.text(r2[i], fy22_values[i] + 10, percentage, ha='center', va='bottom', fontsize=12, fontweight='bold', color=color)

# General layout
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Amount in  Cr', fontsize=14)
plt.title('HealthifyMe Financials FY22', fontsize=16)
plt.xticks([r + bar_width/2 for r in range(len(fy21_values))], categories)
plt.ylim(-200, 450)
plt.legend()

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the plot to a file
plt.tight_layout()
plt.savefig('healthifyme_financials_final_adjusted.png')
plt.show()
```
