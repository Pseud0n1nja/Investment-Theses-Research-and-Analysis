The pie charts depict Bambrew’s expense distribution for FY22 and FY23.
```python
import matplotlib.pyplot as plt

# Data for Bambrew
categories = ['Operating Revenue', 'Total Expenses', 'Profit/Loss']
fy22_values = [9.34, 12.72, -3.38]
fy23_values = [43.52, 62.23, -18.03]
growth_percentages = ['+366%', '+389.2%', '-433.4%']
growth_colors = ['green', 'green', 'red']

# Creating the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Bar width
bar_width = 0.35

# Position of bars on x-axis
r1 = range(len(fy22_values))
r2 = [x + bar_width for x in r1]

# Making the bar plot
bars1 = plt.bar(r1, fy22_values, color='lightseagreen', width=bar_width, edgecolor='grey', label='FY22')
bars2 = plt.bar(r2, fy23_values, color='hotpink', width=bar_width, edgecolor='grey', label='FY23')

# Adding value labels inside the bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval - (0.1 * yval), round(yval, 2), ha='center', va='top', fontsize=10, fontweight='bold', color='black')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval - (0.1 * yval), round(yval, 2), ha='center', va='top', fontsize=10, fontweight='bold', color='black')

# Adding growth percentages above the bars
for i, (percentage, color) in enumerate(zip(growth_percentages, growth_colors)):
    if '-' in percentage:
        plt.text(r2[i], fy23_values[i] - 5, percentage, ha='center', va='top', fontsize=12, fontweight='bold', color=color)
    else:
        plt.text(r2[i], fy23_values[i] + 5, percentage, ha='center', va='bottom', fontsize=12, fontweight='bold', color=color)

# General layout
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Amount in ₹ Cr', fontsize=14)
plt.title('Bambrew Financials FY23', fontsize=16)
plt.xticks([r + bar_width / 2 for r in range(len(fy22_values))], categories)
plt.ylim(-20, 75)
plt.legend()

# Adding grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the plot to a file
plt.tight_layout()
plt.savefig('bambrew_financials_fy23.png')
plt.show()
```

# Financial Metrics FY22 vs FY23

This bar chart showcases key financial metrics for Bambrew over the past two fiscal years. The EBITDA margin slightly worsened from -34% in FY22 to -34.5% in FY23, reflecting the increased costs associated with rapid growth. The expense per rupee of operating revenue marginally increased from ₹1.36 to ₹1.43, indicating a rise in operational efficiency costs. Notably, the Return on Capital Employed (ROCE) plummeted from -27% to -782%, highlighting the significant capital expenditure and the need for improved capital management strategies as the company scales.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
metrics = ["EBITDA Margin", "Expense/₹ of Op Revenue", "ROCE"]
fy22_values = [-34, 1.36, -27]
fy23_values = [-34.5, 1.43, -782]

# Convert percentages to positive values for bar chart
fy22_values_pos = [abs(val) for val in fy22_values]
fy23_values_pos = [abs(val) for val in fy23_values]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Bar width
bar_width = 0.35

# Position of bars on x-axis
r1 = np.arange(len(fy22_values))
r2 = [x + bar_width for x in r1]

# Create bars
bars1 = ax.bar(r1, fy22_values_pos, color='lightseagreen', width=bar_width, edgecolor='grey', label='FY22')
bars2 = ax.bar(r2, fy23_values_pos, color='hotpink', width=bar_width, edgecolor='grey', label='FY23')

# Add value labels above the bars in red
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, f'{-yval}%' if yval == abs(fy22_values[0]) or yval == abs(fy22_values[2]) else f'₹{yval}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='red')

for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, f'{-yval}%' if yval == abs(fy23_values[0]) or yval == abs(fy23_values[2]) else f'₹{yval}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='red')

# Adding details to the plot
ax.set_xlabel('Metrics', fontsize=14)
ax.set_ylabel('Values', fontsize=14)
ax.set_title('FY22 vs FY23 Financial Metrics', fontsize=16)
ax.set_xticks([r + bar_width / 2 for r in range(len(fy22_values))])
ax.set_xticklabels(metrics)
ax.legend()

# Adding grid
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
